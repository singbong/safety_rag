# **Chat API**

```python
import requests
chat_url = "http://203.237.81.58:25723/api/chat"
chat_payload = {
    "question": "지진 발생 시 행동 요령 알려줘",
    "session_id": "user123_session_abc"
}
chat_response = requests.post(chat_url, json=chat_payload)
chat_response.raise_for_status()  # 응답 코드가 200번대가 아니면 에러 발생
chat_result = chat_response.json() # 성공 시 JSON 응답 데이터만 바로 출력
```

```python
print(chat_result.get('generation'))
```

```python
print(chat_result.get('final_answer'))
```

# **generate_form API**

```python
import requests
generate_form_url = "http://203.237.81.58:25723/api/generate_form"
generate_form_payload = {
    "place_name": "2025 한강 여름 뮤직 페스티벌",
    "type": "대규모 야외 공연",
    "region": "서울, 여의도 한강공원",
    "period": "2025년 8월 8일 ~ 2025년 8월 10일",
    "description": "뜨거운 여름밤을 식혀줄 대한민국 최고의 뮤직 페스티벌! 다양한 장르의 아티스트들과 함께하는 3일간의 축제. 푸드트럭 존과 체험 이벤트도 준비되어 있습니다.",
    "category": "음악/페스티벌",
    "related_documents": "전체 타임테이블, 행사장 안내도, 셔틀버스 운행 정보",
    "emergency_contact_name": "종합상황실 안전관리팀",
    "emergency_contact_phone": "02-119-8282" # 예시 연락처
}
generate_form_response = requests.post(generate_form_url, json=generate_form_payload)
generate_form_response.raise_for_status()
generate_form_result = generate_form_response.json()

print(generate_form_result)
```

```python
print(generate_form_result.get("generation"))
```

```python
print(generate_form_result.get("final_answer"))
```

```python
generated_form = generate_form_result.get("final_answer")
```
